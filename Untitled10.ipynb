# Install Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,roc_auc_score, ConfusionMatrixDisplay, roc_curve, auc
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

# Data Visualization

```python
data = pd.read_csv('Cancer_Data.csv')

print(f'We can see that the data shape is: {data.shape}')
```

```python
#this is some rows of the data per se
data.head()
```

```python
#this is technical info from the data
data.info()
```

```python
#we can se how many Null values we have
data.isnull().sum()
```

```python
data.describe()
```

# Data Cleansing

```python
#here we are gonna drop "ID" that is useless and "Unnamed: 32" that is all Null values
data.drop(columns = ['Unnamed: 32', 'id'], axis= 1, inplace= True)
```

```python
#here we are gonna transform the object "diagnosis" to an int, "1" for malignant and "0" for benign
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
```

```python
#correlation for diagnosis in every feature
corr = data.corr()

corr['diagnosis'].sort_values(ascending=False)
```

# Machine learning model

```python
#spliting the data
X = data.drop('diagnosis', axis= 1)
Y = data['diagnosis']

#now we are gona scale the values to make it more simple for the model
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size= 0.2, random_state= 69)

print('The split was successfully done')
```

```python
#training the model
model = LogisticRegression()

model.fit(X_train, Y_train)
```

```python
#predictions
train_score = model.score(X_train, Y_train)
test_score = model.score(X_test, Y_test)

#confusion matrix
predict = model.predict(X_test)
cm = confusion_matrix(predict, Y_test)

#classification report
labels= {0: 'Benign', 1:'Malignant' }
cr = classification_report(predict, Y_test, target_names= labels.values())


#print
print(f'This is the score with train samples: {train_score}')
print(f'This is the score with test samples: {test_score}\n')
print(f'This is the confusion matrix: \n {cm}\n')
print(f'This is the classification report:\n {cr}')
```

```python
# predict probabilities for both train and test data
y_train_pred_prob = model.predict_proba(X_train)[:, 1]
y_test_pred_prob = model.predict_proba(X_test)[:, 1]

# calculate ROC curve for train and test data
fpr_train, tpr_train, thresholds_train = roc_curve(Y_train, y_train_pred_prob)
fpr_test, tpr_test, thresholds_test = roc_curve(Y_test, y_test_pred_prob)

# create subplots to display ROC curves
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# plot ROC curve for train data
plt.sca(axes[0])
plt.plot(fpr_train, tpr_train, linewidth=2, color='teal')
plt.plot([0, 1], [0, 1], 'r--')
plt.title('ROC Curve of train data')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

# plot ROC curve for test data
plt.sca(axes[1])
plt.plot(fpr_test, tpr_test, linewidth=2, color='teal')
plt.plot([0, 1], [0, 1], 'r--')
plt.title('ROC Curve of test data')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

# show the plot
plt.show()
```
